```python
import numpy as np 
import pandas as pd
```

```python
df=pd.read_excel("bike.xlsx")
```

```python
name=[]
for i in df.columns:
    name.append(i.upper())

df.columns=name
```

```python
df.tail(3)
```

```python
df["Young"]=df["AGE_GROUP"].apply(lambda x: 1 if x=="Young Adults (25-34)" else 0)
df["ADULT"]=df["AGE_GROUP"].apply(lambda x: 1 if x=="Adults (35-64)" else 0)
```

```python
df["Young"] = df["Young"].apply(lambda x: "YOUNGER" if x == 1 else 0)

```

```python
df["ADULT"] = df["ADULT"].apply(lambda x: "ADULT" if x == 1 else 0)


```

```python
df["ORDER_QUANTITY"]=df["ORDER_QUANTITY"].apply(lambda x: x.strip(".0"))
```

```python
df.loc[df[" COST "]==0," COST " ]= df["REVENUE"]-df[" PROFIT "]
```

```python
df["custom1"]=df["STATE"]+[" "]+df["COUNTRY"]
```
